2. Explain the purpose of comments in HTML and provide an example of how to use comments in an HTML
document?
  - Comments in HTML are used to add notes or explanations inside the code. These comments are not displayed on the web page in the browser. They are only visible to developers who read the source code.

  # Example of Comments in an HTML Document

```python
<!DOCTYPE html>
<html>
<head>
    <title>HTML Comments Example</title>
</head>
<body>

<!-- This is the main heading of the webpage -->
<h1>Welcome to My Website</h1>

<!-- This paragraph describes the website -->
<p>This website is created to learn HTML.</p>

<!-- The following line is commented and will not appear on the webpage -->
<!-- <h2>This heading will not be displayed</h2> -->

</body>
</html>
```

4. Write a short note on Tag and element with an example?
   - A tag in HTML is a keyword enclosed in angle brackets < > that tells the browser how to display content. Tags usually come in pairs: an opening tag and a closing tag.

       - Opening tag: <p>

       - Closing tag: </p>

    - Tags instruct the browser about structure and formatting of the webpage.

5. What is the DOCTYPE Declaration in HTML?
- The DOCTYPE declaration in HTML is used to tell the web browser which version of HTML is being used in the document. It helps the browser display the web page correctly according to HTML standards.

- It is written at the very top of the HTML document, before the <html> tag.
